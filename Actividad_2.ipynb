```python
import pandas as pd
import numpy as np
import random as random
```

```python
df = pd.read_csv("peso_estatura_genero.csv")
df.Estatura = df.Estatura*2.54
df.Peso = df.Peso/2.2
print(df.info())
```

Existen 3 variables, el género que es cualitativo y la estura y peso que son cuantitativos. Hay 10000 datos disponibles

```python
minEs = df["Estatura"].min()
maxEs = df["Estatura"].max()
print(f"El rango de la estatura es de {minEs} a {maxEs}")

minPes = df["Peso"].min()
maxPes = df["Peso"].max()
print(f"El rango de peso es de {minPes} a {maxPes}")
print("El genero puede ser male o female")
```

Medias

Mediana

```python
df.groupby(["Genero"]).median()
```

Desviacion Estandar

```python
df.groupby(["Genero"]).std()
```

La estatura promedio de las mujeres es menor a la de los hombres al igual que el peso promedio. Los valores centrales de cada dato de los hombres y mujeres se acerca bastante a los valores promedio. Los datos son más dispersos en la estatura de los hombres pero los datos del peso estan dispersos casi de la misma manera.
